```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import statsmodels.api as sm

# Simula√ß√£o de dados (substitua por dados reais)
data = {
    'Ano': [2020, 2021, 2022, 2023],
    'IPI': [73, 80, 90, 95],  # √çndice de Produ√ß√£o Industrial
    'PIB': [2000, 2100, 2200, 2300],  # Produto Interno Bruto
    'Desemprego': [14.7, 12.0, 9.5, 8.0]  # Taxa de Desemprego
}

# Cria√ß√£o do DataFrame
df = pd.DataFrame(data)

# An√°lise descritiva
print("An√°lise Descritiva:")
print(df.describe())

# Gr√°fico da evolu√ß√£o do IPI
plt.figure(figsize=(10, 5))
plt.plot(df['Ano'], df['IPI'], marker='o', label='IPI')
plt.title('Evolu√ß√£o do √çndice de Produ√ß√£o Industrial (IPI)')
plt.xlabel('Ano')
plt.ylabel('IPI')
plt.xticks(df['Ano'])
plt.grid()
plt.legend()
plt.show()

# Gr√°fico da compara√ß√£o do PIB
plt.figure(figsize=(10, 5))
plt.plot(df['Ano'], df['PIB'], marker='o', label='PIB', color='orange')
plt.title('Evolu√ß√£o do Produto Interno Bruto (PIB)')
plt.xlabel('Ano')
plt.ylabel('PIB (em bilh√µes)')
plt.xticks(df['Ano'])
plt.grid()
plt.legend()
plt.show()

# An√°lise de regress√£o
X = df[['IPI', 'Desemprego']]
y = df['PIB']
X = sm.add_constant(X)  # Adiciona uma constante ao modelo

model = sm.OLS(y, X).fit()  # Ajusta o modelo
predictions = model.predict(X)  # Faz previs√µes

# Resultados da regress√£o
print("\nResultados da Regress√£o:")
print(model.summary())

# Gr√°fico de dispers√£o com a linha de regress√£o
plt.figure(figsize=(10, 5))
plt.scatter(df['IPI'], df['PIB'], color='blue', label='Dados Reais')
plt.plot(df['IPI'], predictions, color='red', label='Linha de Regress√£o')
plt.title('Rela√ß√£o entre IPI e PIB')
plt.xlabel('IPI')
plt.ylabel('PIB')
plt.grid()
plt.legend()
plt.show()
```
